```python
from lda2py import LDA
lda = LDA(num_topic=20,max_iter=50)
file_path = "./VietNameseFacebookNovember.txt"
lda.preprocess(file_path)
```

```python
lda.fit()
```

```python
for i in range(len(lda.topic_count)):
    print("Topic",i)
    lda.get_topic_word(no_topic=i,num_word=15)
    print("\n")
```
